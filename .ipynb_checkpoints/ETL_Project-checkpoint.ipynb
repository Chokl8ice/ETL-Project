```python
import pandas as pd
from sqlalchemy import create_engine
```

# CA Law DF

```python
law_csv = "ca_law.csv"
law_df = pd.read_csv(law_csv)
law_df.head()
```

# Clean Data

```python
new_law_df = law_df[['City', 'Total \rofficers', 'Total \rcivilians']].copy()
new_law_df.head()
```

# CA Offenses DF

```python
offenses_csv = "ca_offenses.csv"
offense_df = pd.read_csv(offenses_csv)
offense_df.head()
```

# Clean Data

```python
new_offense_df = offense_df[['City', 'Population', 'Violent crime', 'Murder and nonnegligent manslaughter', 'Rape (revised definition)', 'Robbery', 'Aggravated assault', 'Property crime', 'Burglary', 'Larceny-theft', 'Arson']].copy()
new_offense_df.head()
```

# CA Coordinates DF

```python
coord_csv = "coordinates.csv"
coord_df = pd.read_csv(coord_csv)
coord_df.head()
```

```python
coord_df = coord_df.rename(columns={"Name": "City"})
coord_df.head()
```

# Merging the Data

```python
merged = pd.merge(new_law_df,new_offense_df, on=["City"])
merged.head()
```

```python
clean_merged = merged[['City', 'Population', 'Total \rofficers', 'Total \rcivilians', 'Violent crime', 'Murder and nonnegligent manslaughter', 'Aggravated assault', 'Burglary', 'Arson']].copy()
clean_merged.head()
```

```python
merge2 = pd.merge(clean_merged, coord_df, on=["City"])
merge2.head()
```

```python
final_merge = merge2[['Latitude', 'Longitude', 'City', 'Population', 'Total \rofficers', 'Total \rcivilians', 'Violent crime', 'Murder and nonnegligent manslaughter', 'Aggravated assault', 'Burglary', 'Arson']].copy()
final_merge.head()
```
